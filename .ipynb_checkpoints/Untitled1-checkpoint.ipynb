```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft2, fftshift, fft

# 1D filters
hx = np.array([1, -2, 1])        # Horizontal (x-direction)
hy = np.array([-1, -2, -1])      # Vertical (y-direction)

# Outer product to get 2D filter
h2d = np.outer(hy, hx)

print("2D Filter h[m,n]:")
print(h2d)

# 2D FFT with shift
H_dft = fftshift(fft2(h2d))
H_mag = np.abs(H_dft)

# Plot 3D magnitude spectrum
fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(111, projection='3d')
u = np.arange(3)
v = np.arange(3)
U, V = np.meshgrid(u, v)
ax.plot_surface(U, V, H_mag, cmap='viridis')
ax.set_title("Magnitude Spectrum |H(u,v)|")
ax.set_xlabel("v")
ax.set_ylabel("u")
ax.set_zlabel("|H(u,v)|")
plt.tight_layout()
plt.show()

# Profile H(u, 0): Take center column (n = 1)
H_u0 = np.abs(fft(h2d[:, 1]))
plt.stem(np.arange(3), H_u0)
plt.title("H(u, 0) – Vertical Frequency Profile")
plt.xlabel("u"); plt.ylabel("|H(u, 0)|")
plt.grid(True)
plt.show()

# Profile H(0, v): Take center row (m = 1)
H_0v = np.abs(fft(h2d[1, :]))
plt.stem(np.arange(3), H_0v)
plt.title("H(0, v) – Horizontal Frequency Profile")
plt.xlabel("v"); plt.ylabel("|H(0, v)|")
plt.grid(True)
plt.show()

# Approximate H(u, 1/2): phase shift row-wise by pi (alternate sign)
shifted_col = h2d[:, 0] - h2d[:, 2]  # simulate multiplication by exp(-jπn)
H_u_half = np.abs(fft(shifted_col))
plt.stem(np.arange(3), H_u_half)
plt.title("H(u, 1/2) – Vertical Mid-Frequency Profile")
plt.xlabel("u"); plt.ylabel("|H(u, 1/2)|")
plt.grid(True)
plt.show()

# Approximate H(1/2, v): phase shift column-wise by pi (alternate sign)
shifted_row = h2d[0, :] - h2d[2, :]  # simulate multiplication by exp(-jπm)
H_half_v = np.abs(fft(shifted_row))
plt.stem(np.arange(3), H_half_v)
plt.title("H(1/2, v) – Horizontal Mid-Frequency Profile")
plt.xlabel("v"); plt.ylabel("|H(1/2, v)|")
plt.grid(True)
plt.show()
```
